## **10.6.2 Gradient Descent Implementation**

Gradient Descent is an algorithm which helps to find the slope and intercept of the best fit line to the training data.

```python
m = 0
b = 0
x=vAR_df.iloc[:,0]
y=obj_vAR_df["Final grade_cat"]
L = 0.01 # The learning Rate
epochs = 107  # The number of iterations to perform gradient descent

n = len(vAR_df) # Number of elements in training data
# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*x + b  # The current predicted value of Y
    D_m = (-2/n) * sum(x * (y - Y_pred))  # Derivative wrt m
    D_b = (-2/n) * sum(y - Y_pred)  # Derivative wrt b
    m = m - L * D_m  # Update m
    b = b - L * D_b  # Update b
```

**Red Colour line is the Regression line which is also known as best fit line.**

```python
# Best Fit line with Student id to Categorical value of Grade  
x=vAR_df.iloc[:,0]
y=obj_vAR_df["Final grade_cat"]
vAR_plt.scatter(x, y)
vAR_plt.xlabel('id no.',size=40)
vAR_plt.ylabel('Final_Marks',size=30)
vAR_plt.rc('figure', figsize=(40,60))

vAR_plt.yticks((y),size=20)
m, b = np.polyfit(x, y,1)
print(m)
print(b)
vAR_plt.plot(x, m*x + b,"r")
vAR_plt.rc('figure', figsize=(20,10))
```
