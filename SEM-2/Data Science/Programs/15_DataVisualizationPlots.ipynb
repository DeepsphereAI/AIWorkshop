Line Plot: To create a simple line plot in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y = [2, 4, 6, 8, 10]
plt.plot(vAR_x, vAR_y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Bar Plot: To create a simple bar plot in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y = [2, 4, 6, 8, 10]
plt.bar(vAR_x, vAR_y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Scatter Plot: To create a simple scatter plot in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y = [2, 4, 6, 8, 10]
plt.scatter(vAR_x, vAR_y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Histogram: To create a simple histogram in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt
import numpy as np

vAR_data = np.random.normal(100, 10, 1000)
plt.hist(vAR_data, bins=30)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Pie Chart: To create a simple pie chart in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_labels = ['A', 'B', 'C', 'D']
vAR_values = [20, 30, 40, 10]
plt.pie(vAR_values, labels=vAR_labels, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie Chart')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Box Plot:

```python
import matplotlib.pyplot as plt
import numpy as np

vAR_data = np.random.normal(100, 10, 1000)
plt.boxplot(vAR_data)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Violin Plot:

```python
import matplotlib.pyplot as plt
import numpy as np

vAR_data = np.random.normal(100, 10, 1000)
plt.violinplot(vAR_data)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Violin Plot')

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Stacked Bar Plot:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y1 = [2, 4, 6, 8, 10]
vAR_y2 = [3, 6, 9, 12, 15]
plt.bar(vAR_x, vAR_y1, label='y1')
plt.bar(vAR_x, vAR_y2, bottom=vAR_y1, label='y2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Bar Plot')
plt.legend()

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

vAR_flights = sns.load_dataset("flights")
vAR_flights = vAR_flights.pivot("month", "year", "passengers")

sns.heatmap(vAR_flights, cmap="YlGnBu")

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```

Contour plot:

```python
import matplotlib.pyplot as plt
import numpy as np

def f(vAR_x, vAR_y):
    return np.sin(vAR_x) + np.cos(vAR_y)

vAR_x = np.linspace(-3, 3, 100)
vAR_y = np.linspace(-3, 3, 100)
vAR_X, vAR_Y = np.meshgrid(vAR_x, vAR_y)
vAR_Z = f(vAR_X, vAR_Y)

plt.contour(vAR_X, vAR_Y, vAR_Z)

print("**************************OUTPUT***************************** \n")

plt.show()

print("\n*************************************************************")
```
